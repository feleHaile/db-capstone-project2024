```python
import pandas as pd
import mysql.connector

# Load the Excel data into a DataFrame
file_path = 'C:/Users/Owner/Documents/Meta Database Engineer Course/db-capstone-project2024/Week 4/LittleLemon_Data_Renamed.xlsx'
data = pd.read_excel(file_path)

print(data.columns)

# Connect to the database
conn = mysql.connector.connect(
    host="127.0.0.1",
    user="root",     # Replace with your MySQL username
    password="*******", # Replace with your MySQL password
    database="LittleLemonDBCP"  # Replace with your database name
)
cursor = conn.cursor()

# Function to insert data into Customers table
def insert_customers(data):
    customers = data[['customerID', 'customerName', 'city', 'country', 'postalCode', 'countryCode']].drop_duplicates()
    for index, row in customers.iterrows():
        cursor.execute("""
            INSERT INTO Customers (customerID, customerName, city, country, postalCode, countryCode)
            VALUES (%s, %s, %s, %s, %s, %s)
            ON DUPLICATE KEY UPDATE customerName=VALUES(customerName), city=VALUES(city), country=VALUES(country),
            postalCode=VALUES(postalCode), countryCode=VALUES(countryCode)
        """, (row['customerID'], row['customerName'], row['city'], row['country'], row['postalCode'], row['countryCode']))
    conn.commit()

# Function to insert data into Orders table
def insert_orders(data):
    orders = data[['orderID', 'orderDate', 'deliveryDate', 'customerID', 'cost', 'sales', 'quantity', 'discount', 'deliveryCost']].drop_duplicates()
    for index, row in orders.iterrows():
        cursor.execute("""
            INSERT INTO Orders (orderID, orderDate, deliveryDate, customerID, totalCost, sales, quantity, discount, deliveryCost)
            VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
            ON DUPLICATE KEY UPDATE orderDate=VALUES(orderDate), deliveryDate=VALUES(deliveryDate), 
            customerID=VALUES(customerID), totalCost=VALUES(totalCost), sales=VALUES(sales), 
            quantity=VALUES(quantity), discount=VALUES(discount), deliveryCost=VALUES(deliveryCost)
        """, (row['orderID'], row['orderDate'], row['deliveryDate'], row['customerID'], row['cost'], 
              row['sales'], row['quantity'], row['discount'], row['deliveryCost']))
    conn.commit()

# Function to insert data into OrderItems table
def insert_order_items(data):
    for index, row in data.iterrows():
        cursor.execute("""
            INSERT INTO OrderItems (orderID, courseName, cuisineName, starterName, desertName, drink, sides, quantity, cost)
            VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
        """, (row['orderID'], row['courseName'], row['cuisineName'], row['starterName'], 
              row['desertName'], row['drink'], row['sides'], row['quantity'], row['cost']))
    conn.commit()

# Function to insert data into MenuItems table
def insert_menu_items(data):
    menu_items = data[['courseName', 'cuisineName', 'starterName', 'desertName', 'drink', 'sides']].drop_duplicates()
    for index, row in menu_items.iterrows():
        cursor.execute("""
            INSERT INTO MenuItems (courseName, cuisineName, starterName, desertName, drink, sides)
            VALUES (%s, %s, %s, %s, %s, %s)
            ON DUPLICATE KEY UPDATE cuisineName=VALUES(cuisineName), starterName=VALUES(starterName), 
            desertName=VALUES(desertName), drink=VALUES(drink), sides=VALUES(sides)
        """, (row['courseName'], row['cuisineName'], row['starterName'], row['desertName'], row['drink'], row['sides']))
    conn.commit()

# Function to insert data into Order_MenuItems table
def insert_order_menu_items(data):
    for index, row in data.iterrows():
        # Get the menuItemID from MenuItems table
        cursor.execute("""
            SELECT menuItemID FROM MenuItems 
            WHERE courseName=%s AND cuisineName=%s AND starterName=%s AND desertName=%s AND drink=%s AND sides=%s
        """, (row['courseName'], row['cuisineName'], row['starterName'], row['desertName'], row['drink'], row['sides']))
        menu_item_id = cursor.fetchone()[0]

        cursor.execute("""
            INSERT INTO Order_MenuItems (orderID, menuItemID)
            VALUES (%s, %s)
            ON DUPLICATE KEY UPDATE menuItemID=VALUES(menuItemID)
        """, (row['orderID'], menu_item_id))
    conn.commit()

# Function to insert data into Bookings table
def insert_bookings(data):
    # Assuming bookingID and bookingDate are derived or provided elsewhere
    bookings = data[['customerID']].drop_duplicates()
    for index, row in bookings.iterrows():
        # Example bookingID and bookingDate generation (should be replaced with real data)
        booking_id = index + 1
        booking_date = '2023-01-01'  # Placeholder date, replace with actual booking date
        cursor.execute("""
            INSERT INTO Bookings (bookingID, bookingDate, customerID)
            VALUES (%s, %s, %s)
        """, (booking_id, booking_date, row['customerID']))
    conn.commit()

# Function to insert data into OrderDeliveryStatus table
def insert_order_delivery_status(data):
    delivery_status = data[['orderID', 'deliveryDate']].drop_duplicates()
    for index, row in delivery_status.iterrows():
        status = 'Delivered'  # Placeholder status, replace with actual status
        cursor.execute("""
            INSERT INTO OrderDeliveryStatus (orderID, status, deliveryDate)
            VALUES (%s, %s, %s)
            ON DUPLICATE KEY UPDATE status=VALUES(status), deliveryDate=VALUES(deliveryDate)
        """, (row['orderID'], status, row['deliveryDate']))
    conn.commit()

# Insert data into tables
insert_customers(data)
insert_orders(data)
insert_menu_items(data)
insert_order_items(data)
insert_order_menu_items(data)
insert_bookings(data)
insert_order_delivery_status(data)

# Close the connection
cursor.close()
conn.close()
```
